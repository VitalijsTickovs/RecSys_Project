### Check average rating per category

```python
import numpy as np
```

```python
games_df = pd.read_csv(data_folder + 'games.csv')
```

```python
thematic_games = games_df[games_df['Cat:Thematic'] == 1]
strategy_games = games_df[games_df['Cat:Strategy'] == 1]
war_games = games_df[games_df['Cat:War'] == 1]
family_games = games_df[games_df['Cat:Family'] == 1]
cgs_games = games_df[games_df['Cat:CGS'] == 1]
abstract_games = games_df[games_df['Cat:Abstract'] == 1]
party_games = games_df[games_df['Cat:Party'] == 1]
childrens_games = games_df[games_df['Cat:Childrens'] == 1]
```

```python
cat_sizes = []

cats_dfs = [thematic_games,strategy_games,war_games,family_games,cgs_games,abstract_games,party_games,childrens_games]

for cat in cats_dfs:
    cat_sizes.append(cat.shape[0])
```

```python
cat_names = ['Thematic','Strategy','War','Family','CGS','Abstract','Party','Childrens']
```

```python
fig = plt.figure(figsize = (10, 5))

plt.bar(cat_names, cat_sizes,
        width = 0.5)
 
plt.xlabel("Categories")
plt.ylabel("# of games")
plt.title("Number of games in different categories")
plt.show()
```

```python
cat_ratings_mean = []

for cat in cats_dfs:
    cat_ratings_mean.append(np.mean(cat['AvgRating']))
```

```python
fig = plt.figure(figsize = (10, 5))

plt.bar(cat_names, cat_ratings_mean,
        width = 0.5)
 
plt.xlabel("Categories")
plt.ylabel("mean rating")
plt.title("Mean rating of games in different categories")
plt.show()
```

```python
cat_ratings_median = []

for cat in cats_dfs:
    cat_ratings_median.append(np.median(cat['AvgRating']))
```

```python
fig = plt.figure(figsize = (10, 5))

plt.bar(cat_names, cat_ratings_median,
        width = 0.5)
 
plt.xlabel("Categories")
plt.ylabel("median rating")
plt.title("Median rating of games in different categories")
plt.show()
```
